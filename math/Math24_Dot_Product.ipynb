<i>Скалярное произведение</i> - это умножение между двумя одинаковыми по размерности векторами. 

Так он называется, потому что результатом умножения будет <i>скалярная величина</i>.

Рассмотрим два вектора:
$$
    u = \myrvector{-3 \\ -2 \\ 0 \\ -1 \\ 4} \mbox{ и } v = \myrvector{-1\\ -1 \\2 \\ -3 \\ 5}.
$$

Скалярное произведение $ u $ и $ v $, обозначаемое как $ \dot{u}{v}$, можно задать алгоритмически.

<u>Попарное умножение</u> - это умножение значений векторов в одинаковых позициях друг на друга.

Результатом скалярного произведения будет сумма всех результатов, полученных в ходе попарных умножений.


```python
# Инициализируем два вектора
u = [-3,-2,0,-1,4]
v = [-1,-1,2,-3,5]

uv = 0;

for i in range(len(u)):
    print("Результат попарного умножения:",u[i]*v[i])
    uv = uv + u[i]*v[i] # сумма всех результатов
    
print() # вывод пустой строки   
print("Скалярное произведение",u,'и',v,uv)
```

Попарное умножение элементов векторов u и v
<ul>
    <li> $ (-3)\cdot(-1) = 3 $, </li>
    <li> $ (-2)\cdot(-1) = 2 $, </li>
    <li> $ 0\cdot 2 = 0 $, </li>
    <li> $ (-1)\cdot(-3) = 3 $, и, </li>
    <li> $ 4 \cdot 5  = 20 $. </li>
</ul>

Таким образом, сумма всех попарных умножений записей равна $ 3+2+0+3+20 = 28 $.

<b>Отметим, что размерности этих векторов должны быть равны. В противном случае скалярного произведения не произойдет.</b>

```python
# найдем скалярное произведение векторов v и u
v = [-4,0]
u = [0,-5]
result = 0;

for i in range(2):
    result = result + v[i]*u[i]

print("Скалярное произведение u и v равна",result)
```

```python

v = [-4,3]
u = [-3,-4]
result = 0;

for i in range(2):
    result = result + v[i]*u[i]

print("Скалярное произведение u и v равна",result)
```

Скалярное произведение между новыми векторами $ u $ и $ v $ также равно $0$. 

Это неудивительно, потому что векторы $u$ и $v$ (в обоих случаях) перпендикулярны.

<ul>
    <li>Скалярное произведение двух ортогональных (перпендикулярных) векторов равно нулю.</li>
    <li>Если скалярное произведение двух векторов равно нулю, то они ортогональны друг другу.</li>
</ul>

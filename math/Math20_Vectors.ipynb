<b>Вектор</b> - это массив чисел. 
Они очень нужны для описания состояния квантовой системы.
Вектор - это математический объект, который мы можем представить в виде массива.
Количество элементов в массиве - это его длина.


```python
# рассмотрим следующий список из 4 элементов 
L = [1,-2,0,5]
print(L)
```

Векторы могут иметь горизонтальную или вертикальную форму:
$$
    u = \mypar{1~~-2~~0~~-5} ~~~\mbox{ или }~~~ v =\mymatrix{r}{1 \\ -2 \\ 0 \\ 5}, ~~~\mbox{ соответственно.}
$$
Отметим, что в векторном представлении нам не нужно использовать запятую.

<h3> Умножение вектора на число</h3>
Каждый элемент массива умножается на это число.
$$
    3 \cdot v = 3 \cdot  \mymatrix{r}{1 \\ -2 \\ 0 \\ 5} = \mymatrix{r}{3 \\ -6 \\ 0 \\ 15}
    ~~~~~~\mbox{ или }~~~~~~
    (-0.6) \cdot v = (-0.6) \cdot \mymatrix{r}{1 \\ -2 \\ 0 \\ 5} = \mymatrix{r}{-0.6 \\ 1.2 \\ 0 \\ -3}.
$$


```python
# 3 * v
v = [1,-2,0,5]
print("v is",v)
# мы используем тот же массив для результата
for i in range(len(v)):
    v[i] = 3 * v[i]
print("3v is",v)

# -0.6 * u
# переинициализируем список v
v = [1,-2,0,5]
for i in range(len(v)):
    v[i] = -0.6 * v[i]
print("0.6v is",v)
```

<h3> Сложение векторов</h3>

При сумме двух векторов числа на одинаковых позициях складываются.

$$
    u = \myrvector{-3 \\ -2 \\ 0 \\ -1 \\ 4} \mbox{ и } v = \myrvector{-1\\ -1 \\2 \\ -3 \\ 5}.
    ~~~~~~~ \mbox{}~~
    u+v = \myrvector{-3 \\ -2 \\ 0 \\ -1 \\ 4} + \myrvector{-1\\ -1 \\2 \\ -3 \\ 5} =
    \myrvector{-3+(-1)\\ -2+(-1) \\0+2 \\ -1+(-3) \\ 4+5} = \myrvector{-4\\ -3 \\2 \\ -4 \\ 9}.
$$

В Python мы можем сделать те же вычисления.

```python
u = [-3,-2,0,-1,4]
v = [-1,-1,2,-3,5]
result=[]
for i in range(len(u)):
    result.append(u[i]+v[i])

print("u+v is",result)

# выводим результат
print() # выводим пустую строку
print("элементы вектора u+v:")
for j in range(len(result)):
    print(result[j])
```

<h3> Длина вектора </h3>

Длина вектора - это кратчайшее расстояние от начала координат $(0,0)$ до точки, в которой этот вектор заканчивается.

Длину вектора можно вычислить с помощью теоремы Пифагора. 

Рассмотрим вектор $ u = \myrvector{-3 \\ 4} $.

```python
%run math.py
visualize_vectors("example3")
```

Длина $ u $ обозначается как $ \norm{u} $, и вычисляется как $ \norm{u} =\sqrt{(-3)^2+4^2} = 5 $. 
$$
    v = \myrvector{-1 \\ -3 \\ 5 \\ 3 \\ 1 \\ 2}
    ~~~~~~~~~~
    \mbox{и}
    ~~~~~~~~~~
    \norm{v} = \sqrt{(-1)^2+(-3)^2+5^2+3^2+1^2+2^2} .
$$
Найдем длину следующего вектора с помощью Python:

<div style="font-style:italic;background-color:#fafafa;font-size:10pt;">    <ul>
        <li>В общем виде: $ a^x $ в Python можно обозначить как $ a ** x $. </li>
        <li>Квадрат числа $a$: $ a^2 $ можно обозначить как $ a ** 2 $ в Python.</li>
        <li> Квадратный корень из числа $a$: $ \sqrt{a} = a^{\frac{1}{2}} = a^{0.5} $ можно обозначить как $ a ** 0.5 $ в Python.</li>
    </ul>
</div>

```python
v = [-1,-3,5,3,1,2]

length_square=0
for i in range(len(v)):
    print(v[i],":square ->",v[i]**2) # выводим каждую запись и ее квадратное значение
    length_square = length_square + v[i]**2 # суммируем квадрат каждого элемента

length = length_square ** 0.5 # извлекаем квадратный корень из суммы квадратов всех элементов
print("Сумма равна",length_square)
print("Длина равна",length)

#Для получения квадратного корня можно также использовать встроенную функцию math.sqrt
print() # выводим пустую строку
from math import sqrt
print("квадратный корень из",length_square,"равен",sqrt(length_square))
```

<h3> Примечания:</h3>

Рассмотрим вектор $ -3 v $. Он имеет ту же длину, что и $ 3v $, но его направление противоположно.

Поэтому, вычисляя длину $ -3 v $, мы используем модуль числа:

$ \norm{-3 v} = |-3| \norm{v} = 3 \norm{v}  $.

Здесь $ |-3| $ - это модуль $ -3 $.

Модуль - это его расстояние до 0. Таким образом, $ |-3| = 3 $.

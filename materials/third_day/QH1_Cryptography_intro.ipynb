# Квантовые вычисления и криптография

Квантовые вычисления используются в разных областях, одной из которых является криптография.

## Криптография

Криптография — это наука о шифровании данных. Сегодня она широко применяется для защиты банковских операций, обмена сообщениями в мессенджерах, электронной почты и других цифровых коммуникаций.

### Пример обеспечения безопасности информации

Чтобы лучше понять понятие безопасности информации, рассмотрим следующий пример. Алиса и Боб имеют сундук со сладостями. Сундук оснащен замком, ключи от которого имеются только у Алисы и Боба, что обеспечивает *конфиденциальность* — посторонние лица не могут открыть сундук.

### Функции криптографии

- **Конфиденциальность:** Безопасный обмен сообщениями, предотвращающий доступ посторонних лиц к информации.
- **Аутентификация:** Возможность получателя проверить подлинность отправителя сообщения.
- **Целостность данных:** Гарантия того, что сообщение не было изменено или повреждено при передаче.
- **Неотрицание авторства:** Отправитель не может отрицать факт отправки сообщения.

Для этого используются различные методы, алгоритмы и протоколы.

### Хеш-функции

Одним из таких инструментов являются хеш-функции. Хеш-функции — это односторонние функции, создающие короткие уникальные "отпечатки" больших данных.

Название "односторонние" объясняется тем, что легко получить хеш из данных, но невозможно восстановить исходные данные или создать разные входы с одним и тем же хешем.

Хеш-функции используются для проверки целостности данных. Например, если вы скачали книгу из Интернета, хеш-значение книги, предоставляемое автором, поможет проверить, что файл не был изменён или повреждён.

Хеш-функции активно используются в различных криптографических протоколах, для формирования цифровой подписи, т.к. удобно подписывать хеш-значение данных, для которых требуется подпись.


### Задача: сравнение двух строк

Пусть имеются две строки одинаковой длины $s$ и $t$, состоящие только из нулей и единиц. Необходимо проверить равенство этих бинарных строк, т.е. выполняется ли соотношение $s = t$. Примеры:

$$
s=0010101 , t=0010101 \rightarrow s=t,
$$

$$
s=0010101 , t=0011101 \rightarrow s \neq t.
$$

Вместо того, что непосредственно сравнивать исходные строки, будем сравнивать их краткие образы — хеши.

Через $h(x)$ обозначим хеш-функицию. Для строк $s$ и $t$ вычисляем $h(s)$ и $h(t)$ и сравниваем полученные хеш-значения.

Если $h(s) \neq h(t)$, то строки $s$ и $t$ различны.

Если $h(s)=h(t)$, то с высокой вероятностью строки идентичны.

Так как мы отображаем длинный вход в короткий выход, то можем столкнуться с ситуацией, когда две разных строки имеют одно и то же хеш-значение:

$$
s \neq t
$$
такие, что
$$
h(s)=h(t).
$$.

Такую ситуацию принято называть *коллизией*.

### Алгоритм сравнения строк с использованием хеширования

Выберем простое число $p$ и определим хеш-функцию $h$ следующим образом:

$$
h(x)=\#_1(x)\bmod p,
$$

где $\#_1(x)$ - количество единиц в строке $x$.

Запись $\bmod p$ читается как "по модулю $p$" и означает, что берется остаток от деления на $p$.
Например,

$$
7 \bmod 3 = 1,
$$
$$
9 \bmod 7 = 2.
$$

### Пример

Пусть $p = 3$.

$s=0110101$, $\#_1 (s) = 4$, $h(s)= 4 \bmod 3 = 1$.

$t=0110101$, $\#_1 (t) = 4$, $h(t)= 4 \bmod 3 = 1$. 

Так как $h(s) = h(t)$, то получаем, что строки равны.

### Улучшенный алгоритм для сравнения строк

Для того, чтобы уменьшить вероятность коллизий, будем брать не одно число $p$, а некоторый набор $P = \{p_1, p_2, ..., p_n\}$ и определим $n$ хеш-функций:

$$
h_1(x) = \#_1(x) \bmod p_1,
$$
$$
h_2(x) = \#_1(x) \bmod p_2,
$$
$$
...
$$
$$
h_n(x) = \#_1(x) \bmod p_n.
$$

Теперь будем считать, что строки $s$ и $t$ равны только в том случае, если $h_i(s) = h_i(t)$ для всех индексов $i \in [1, n]$. Если хотя бы для одного индекса $i$ $h_i(s) \neq h_i(t)$, то строки различны.

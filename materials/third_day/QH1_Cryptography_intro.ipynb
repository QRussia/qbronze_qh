Квантовые вычисления имеют разные области применения. Один из таких областей - криптография.

Криптография - наука шифрования.

В наши дни криптография имеет широкое применение - безопасность банковских операций, обмен сообщениями в мессенджерах, электронная почта и т.д. и т.п.

### Функции криптографии

- Обеспечение конфиденциальности (безопасный обмен сообщения).
- Проверка подлинности (аутентификация) (получатель сообщения может проверить его источник).
- Целостность (проверка сообщения на то, что оно не было изменено в процессе передачи).
- Неотрицание авторства (отправитель не сможет ложно отрицать отправку сообщения).

Для реализации этих функций разработаны различные методы, алгоритмы и протоколы.

### Функции хэширования

Один из таких инструментов - односторонние функции хэширования (или просто **хэш-функции**).

Хэш-функции напоминают цифровые отпе­чатки пальцев (Fingerprinting): небольшие фрагменты данных, которые могут служить для иден­тификации достаточно больших цифровых объектов. Это общедоступные функ­ции, у них нет никаких секретных ключей.

Они названы односторонними из-за своей математической природы. Любой может вычислить одностороннее хэш-значение чего угодно (например, текста книги). Однако если имеется хэш-значение конкретной книги, исходя из вычислений не­ возможно создать другую книгу с таким же значением хэш-функции или полу­чить подлинный текст книги.

Хэш-функция также может обеспечивать аутентификацию и целостность. Рассмотрим также на примере книги. 

Если бы вы загрузили книгу из Интернета, у вас не было бы никакого способа узнать, был ли это оригинал, или кто-то другой все же частично изменил книгу. Однако, если бы автор книги дал вам в руки хэш-значение для нее, вы смогли бы сравнить расчетный результат с этим значением. Если они совпадают, то у вас оригинал, без изменений.

Хэш-функции широко применяются в криптографии и компьютерной безопас­ности. Они используются почти во всех протоколах Интернета, чтобы обрабаты­вать ключи, связывать последовательность событий или аутентифицировать со­бытия. Они также важны для алгоритмов цифровой подписи.

### Формальная задача

Рассмотрим пример и на основе него сформируем задачу. Для простоты будем рассматривать сравнивание бинарных строк.

Есть две бинарные строки $s$ и $t$. Надо проверить $s = t$ или нет.

$$
s=0010101 , t=0010101
$$

$$
s=0010101 , t=0011101
$$

Самый простой алгоритм проверки - побитовая проверка. Однако, это долго и не удобно.

Для проверки равенства строк можем воспользоваться хэш-функцией.

Введем обозначение хэш-функиции $h(x)$.

Для строк $s$ и $t$ считаем $h(s)$ и $h(t)$ и сравниваем полученные значения.

Если $h(t)=h(s)$, то говорим, что строки равны, если нет, то не равны.

Так как множество потенциально сравниваемых строк бесконечно, а множество значений хэш-функций конечно, то могут быть случаи, когда строки не равны, но полученные хэши - равны. Это называется **коллизия**.

Однако, вопрос, 

### как построить хэш-функцию?

Требования, которые мы накладываем на хэш-функцию:
- сравнение результата хэш-функции (или хэша) должна быть проще, чем сравнивать сами данные;
- хэш-функция должна работать быстро;
- меньше коллизии.

Существует много алгоритмов для построения хэш-функций.

Мы рассмотрим несколько алгоритмов, на которых будет базироваться квантовый алгоритм построения хэш-функции.

### Детерминированный алгоритм

Берем простое число $p$.

$h(x)=\#_1(x) mod p$, где $\#_1(x)$ - количество $1$ в строке, $mod p$ - по модулю $p$.

### Пример

Пусть $p = 3$.

$s=0110101$, $\#_1 (s) = 4$, $h(s)=1$.

$t=0110101$, $\#_1 (t) = 4$, $h(t)=1$. 

Получаем, что - строки равны.

Для борьбы с коллизией можно придумывать различные решения.

Например, вместо $\#_1(x)$ можно взять другую функцию, как вариант, $val(x)$ - 10-чное значение бинарной строки.

### Рандомизированный алгоритм

Усилия данного алгоритма направлены на то, чтобы уменьшить коллизию с помощью более умного выбора простого числа $p$. Для этого берется некоторое количество простых чисел, и среди них равновероятностно выбирается число $p$. 

В качестве дополнительного улучшения можно взять прогнать алгоритм несколько раз для разных $p$. Тогда, две строки будут равны в том случае, если хэши на каждом из прогонов будут равны. Иначе строки различны.

<h2>Магические команды</h2>

Здесь мы перечислим некоторые встроенные магические команды Jupyter ноутбуков, которые мы будем использовать в нашем материале. 

Эти команды могу быть выполнены в ячейках с типом code.

<b><u>Записать содержимое ячейки с типом code</u></b> во внешний файл:
    
    %%writefile FILENAME.py
Эта команда должна быть расположена в первой строке ячейки, далее нужно выполнить эту ячейку.

<i> Пример:</i>

```python
%%writefile first.py 

print("hello world")

str="*"
for i in range(5):
    print(str)
    str+="*"  
```

<b><u>Выполнение внешнего скрипта</u></b> без загрузки его содержимого в ячейку:
    
    %run FILENAME.py
<i>Пример:</i>    

```python
%run first.py
```

<b><u>Загрузка внешнего скрипта</u></b> в ячейку:

    %load FILENAME.py
Как только команда будет выполнена, содержимое ячейки будет заменено на содержимое файла. (Предыдущее содержимое будет удалено.)

Кроме того, эта команда будет расположена на первой строке и закомментирована.

<i>Пример:</i>

```python
%load first.py
```
